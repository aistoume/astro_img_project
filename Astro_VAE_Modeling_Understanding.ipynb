```python
%matplotlib inline
# %matplotlib notebook
def visualize_kernels(kernels, title="", max_channels=16):
    """
    kernels: Tensor of shape (out_channels, in_channels, kH, kW)
    """
    kernels = kernels[:max_channels]  # 
    num_kernels = kernels.shape[0]
    in_channels = kernels.shape[1]
    kH, kW = kernels.shape[2], kernels.shape[3]

    fig, axs = plt.subplots(1, num_kernels, figsize=(num_kernels, 2))
    if num_kernels == 1:
        axs = [axs]
    for i in range(num_kernels):
        kernel = kernels[i].mean(dim=0).detach().cpu()  #  in_channels 
        axs[i].imshow(kernel, cmap='gray')
        axs[i].axis('off')
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()


# 
model = vae  # 
model.eval()

#  Encoder  Decoder 
for name, module in model.encoder.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        print(f"[Encoder] Visualizing: {name}")
        visualize_kernels(module.weight.data.clone(), title=f"Encoder - {name}")

for name, module in model.decoder.named_modules():
    if isinstance(module, torch.nn.ConvTranspose2d):
        print(f"[Decoder] Visualizing: {name}")
        visualize_kernels(module.weight.data.clone(), title=f"Decoder - {name}")
```

```python
def extract_latent_features(model, input_image):
    """
    model - AstroConvVae 
         input_image - , Tensor (1, 1, 128, 128)
     dict fc_mu, fc_logvar, fc_bias, fc_cf, fc_decode 
    """
    model.eval()
    with torch.no_grad():
        # 
        encoded = model.encoder(input_image)  # shape: (1, 128, 16, 16)
        flat = encoded.view(1, 131072)  # flatten to (1, 131072)
        # flat = encoded.view(-1, encoded.size(0))  # flatten to (131072, 1)
        
        # 
        mu = model.fc_mu(flat)
        logvar = model.fc_logvar(flat)
        bias = model.fc_bias(flat)
        cf = model.fc_cf(flat)
        decode_vector = model.fc_decode(mu)  #  mu

        return {
            "fc_mu": mu,
            "fc_logvar": logvar,
            "fc_bias": bias,
            "fc_cf": cf,
            "fc_decode": decode_vector
        }
```
